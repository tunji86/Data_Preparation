pd.concat() Method

```python
x = pd.concat([DFweek1,DFweek2],join='inner',axis=0, keys=['week1','week2'])
```

```python
x.loc['week1']
```

```python
DFweek2.append(DFweek1, ignore_index=True, sort=True)
```

.merge() method

```python
DFweek1.merge(DFweek2,how = 'outer', on='Customer ID', suffixes=[' W1',' W2'])
```

```python
DFweek1.merge(DFweek2,how = 'left', on=['Customer ID','Food ID'], suffixes=[' W1',' W2'])
```

```python
mask = DFweek1.merge(DFweek2,how = 'outer', on='Customer ID', suffixes=[' W1',' W2'],indicator=True)
```

```python
mask[mask['_merge'].isin(['left_only','right_only'])]
```

```python
DFweek1.merge(DFfood, on='Food ID',how='right',sort=True)
```

```python
DFweek2.merge(DFCust,how='inner',left_on='Customer ID',right_on='ID').drop('ID',axis=1)
```

    MERGING BY INDEXES WITH THE left_index and right_index PARAMETERS #right_index=True or left_index=True

```python
DFweek1.merge(DFCust, how='left',left_on='Customer ID',right_index=True)
```
