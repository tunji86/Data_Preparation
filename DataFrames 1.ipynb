Null values

```python
#Dataframe.dropna(how="all")-->it drops records with all values Null when 'how' is set to "all", and drops 
#records with any null values when 'how' is set to "any". 'Subset' specifies that we only want records dropped
#if that column is Null
nba.dropna(how='all')
```

Replace Null Values, .fillna()

```python
#Dataframe["column"].fillna("New Value")
nba["College"].fillna("No college",inplace=True)#All null values in College column now replaced with "No college"
nba["Salary"].fillna("0", inplace=True)
nba["Tax"].fillna("0", inplace=True)
```

```python
nba.head(5)
```

The .astype() Method--> For changing Series(row in Dataframe) datatypes

```python
#Dataframe["column"].astype("new_type")
nba["Salary"] = nba["Salary"].astype("int")
```

category datatype

```python
#stores column values in memory and replaces them with pointers to their value's memory location, therefore reducing 
#memory usage and increasing speed
nba["Position"] = nba["Position"].astype("category")
```
