```python
import pandas as pd
```

```python
emp = pd.read_csv("employees.csv", parse_dates=(["Start Date","Last Login Time"]))
emp.head()
```

```python
emp.info()
```

```python
emp["Gender"] = emp["Gender"].astype("category")
```

```python
emp["Start Date"] = pd.to_datetime(emp["Start Date"])
```

```python
emp["Last Login Time"] = pd.to_datetime(emp["Last Login Time"])
```

Filter a DF based on a condition

```python
emp[emp["Gender"] == "Male"]
```

```python
mask = emp["Gender"] == "Male"
emp[mask].head()
```

```python
mask2 = emp["Gender"] != "Male"
mask2.head()
```

```python
emp["Before 85"] = emp["Start Date"] <= "1985-01-01"
emp.head()
```

Filter with multiple conditions

```python
gender_mask = emp["Gender"] == "Male"

Marketing_mask = emp["Team"] == "Marketing"

dt1_mask = emp["Salary"] > 10000

emp[(gender_mask & Marketing_mask) | dt1_mask]
```

```python
SM_mask = emp["Senior Management"] #no condition means all records pass
dt_mask = emp["Salary"] > 10000

emp[SM_mask | dt_mask] 
```

The .duplicate() method --> It extracts duplicate rows

```python
emp.sort_values("First Name", inplace=True)
```

```python
var_dup =  ~emp["First Name"].duplicated(keep = False)#returns only duplicate records of specified column(s)
emp[var_dup]
```

The .drop_duplicates() Method

```python
s = emp.drop_duplicates(subset="First Name")
s
```
