Grouping by Multiple Columns

```python
g = DFFort.groupby(['Sector','Industry'])
```

```python
g['Profits','Employees'].mean()
```

.agg() Method #can be called on a groupby() variable

```python
g.agg(('sum','mean'))
g.agg({'Revenue':'sum',
        'Profits':'mean'})
```

```python
g['Location','Profits'].max()
```

```python
DFFort.nlargest(1,'Profits')
```

```python
df = pd.DataFrame(columns=DFFort.columns)
```

```python
for city, data in g:
    highest_city_revenue = data.nlargest(1, 'Profits')
    df = df.append(highest_city_revenue)
```
